# Outbound Auth

Outbound Auth는 agent와 AgentCore Gateway가 Inbound Auth 중에 인증 및 권한이 부여된 사용자를 대신하여 AWS 리소스 및 타사 서비스에 안전하게 액세스할 수 있도록 합니다. AWS 리소스 또는 타사 서비스와 권한 부여를 통합하려면 Inbound Auth와 Outbound Auth를 모두 구성해야 합니다.

AgentCore Identity가 지원하는 최소 권한 액세스 및 안전한 권한 위임을 통해 agent는 GitHub, Google, Salesforce, Slack과 같은 AWS 리소스 및 타사 도구에 원활하고 안전하게 액세스할 수 있습니다. Agent는 사용자를 대신하여 또는 사전 승인된 사용자 동의가 있는 경우 독립적으로 이러한 서비스에서 작업을 수행할 수 있습니다. 또한 안전한 토큰 저장소를 사용하여 동의 피로를 줄이고 간소화된 AI agent 경험을 만들 수 있습니다.

## Outbound Authentication 구성

먼저 타사 provider에 클라이언트 애플리케이션을 등록한 다음 Outbound Auth를 생성합니다. AWS 리소스 또는 타사 서비스 또는 AgentCore Gateway 대상에 대한 액세스를 검증하는 방법을 지정합니다. OAuth 2LO/3LO 또는 API 키를 사용할 수 있습니다. OAuth를 사용하면 AgentCore Identity가 제공하는 provider 중에서 선택할 수 있습니다. 이 경우 AgentCore Identity에서 provider에 대한 구성 세부 정보를 입력합니다. 또는 커스텀 provider에 대한 세부 정보를 제공할 수 있습니다. 

사용자가 AWS 리소스 또는 타사 서비스 또는 AgentCore Gateway 대상에 액세스하려는 경우 Outbound Auth는 Incoming Auth에서 제공한 액세스 토큰이 유효한지 확인하고 유효한 경우 리소스에 대한 액세스를 허용합니다.

## Resource credential provider

이것은 agent가 다운스트림 리소스 서버(예: Google, GitHub)의 자격 증명을 검색하여 액세스하는 데 사용하는 구성 요소입니다(예: Github에서 비공개 저장소 가져오기). 최종 사용자, agent 코드 및 외부 권한 부여 서버 전반에 걸쳐 2LO 및 3LO OAuth2 오케스트레이션 플로우를 구현하는 agent 개발자의 부담을 덜어줍니다. AgentCore는 커스텀 OAuth2 credential provider와 권한 부여 서버 엔드포인트 및 provider별 매개변수가 미리 채워진 Google, GitHub, Slack, Salesforce와 같은 내장 provider 목록을 모두 제공합니다.
  

Bedrock AgentCore Identity는 agent 개발자가 OAuth2 또는 API 키를 지원하는 외부 리소스로 인증할 수 있도록 OAuth2 및 API Key Credential Provider를 제공합니다. 다음 예제에서는 API Key credential provider 구성 과정을 안내합니다. 그런 다음 agent는 API Key credential provider를 사용하여 모든 agent 작업에 대한 API 키를 검색할 수 있습니다. 다른 credential provider에 대해서는 문서를 참조하십시오.

## OAuth2 Authorization URL Session Binding 프로세스

OAuth2 authorization URL session binding 프로세스는 OAuth2 권한 부여 세션이 AgentCore Identity에서 인증된 사용자와 올바르게 연결되도록 하는 중요한 보안 메커니즘입니다. 이 프로세스는 세션 하이재킹을 방지하고 OAuth 토큰이 의도된 사용자에게만 부여되도록 합니다.

참고: https://docs.aws.amazon.com/bedrock-agentcore/latest/devguide/oauth2-authorization-url-session-binding.html

### session binding 작동 방식
<div style="text-align:center">
    <img src="images/identity-session-binding.png" width="90%"/>
</div>

1. agent 호출 – agent 코드가 GetResourceOauth2Token API를 호출하여 권한 부여 URL을 검색합니다. 원래 agent 사용자가 자신이 소유한 일부 애플리케이션 또는 리소스에 액세스하려는 경우입니다.

2. 권한 부여 URL 생성 – AgentCore Identity는 사용자가 탐색하고 액세스에 동의할 수 있도록 권한 부여 URL 및 세션 URI를 생성합니다.

3. 권한 부여 및 액세스 토큰 획득 – 사용자가 권한 부여 URL로 이동하여 agent가 자신의 리소스에 액세스하는 것에 동의합니다. 그 후 AgentCore Identity는 사용자의 브라우저를 권한 부여 요청의 원래 사용자에 대한 정보가 포함된 HTTPS 애플리케이션 엔드포인트로 리디렉션합니다. 이 시점에서 HTTPS 애플리케이션 엔드포인트는 원래 agent 사용자가 애플리케이션의 현재 로그인한 사용자와 동일한지 확인합니다. 일치하면 애플리케이션 엔드포인트가 CompleteResourceTokenAuth를 호출하여 AgentCore Identity가 액세스 토큰을 가져와 저장할 수 있도록 합니다.

4. agent를 다시 호출하여 액세스 토큰 획득 – 애플리케이션이 유효한 응답을 반환하면 agent 애플리케이션은 사용자를 위해 원래 요청된 OAuth2.0 액세스 토큰을 검색할 수 있습니다. 사용자가 일치하지 않으면 애플리케이션은 아무 작업도 수행하지 않거나 시도를 기록합니다.

애플리케이션 엔드포인트가 사용자 ID를 확인할 수 있도록 함으로써 AgentCore Identity는 agent 애플리케이션이 항상 권한 부여 요청을 시작한 사용자와 액세스에 동의한 사용자가 동일한지 확인할 수 있도록 합니다.

### 이 샘플의 OAuth2 Session Binding 플로우 개요

OAuth2 session binding 프로세스에는 애플리케이션, AgentCore Identity, 외부 OAuth provider(Github, Google 등) 및 로컬 콜백 서버 간의 조정이 포함된 여러 주요 단계가 있습니다:

#### 1단계: Application Callback URL 생성
- 애플리케이션에서 공개적으로 액세스 가능한 HTTPS 콜백 엔드포인트를 생성합니다
- 이 엔드포인트는 OAuth 리디렉션을 처리하고 사용자 세션을 검증합니다
- 예시: `https://myagentapp.com/callback`

#### 2단계: Callback URL로 Workload Identity 업데이트
- workload identity에 콜백 URL을 `AllowedResourceOauth2ReturnUrl`로 등록합니다
- 이는 `UpdateWorkloadIdentity` API를 사용하여 수행됩니다
- **이 튜토리얼에서**: 이 단계는 환경 인식 콜백 서버 URL로 workload identity를 업데이트하는 아래 코드에 의해 자동으로 처리됩니다

#### 3단계: OAuth2 Credential Provider 생성
- 외부 OAuth provider 세부 정보(client ID, secret)로 credential provider를 구성합니다
- AgentCore Identity는 provider에 대한 고유한 콜백 URL을 반환합니다
- 외부 OAuth provider(예: GitHub Console)에 이 콜백 URL을 등록합니다

#### 4단계: Session 검증 및 Token 완료 구현
- 콜백 엔드포인트는 현재 사용자의 세션을 검증해야 합니다
- 사용자 식별자 및 세션 URI로 `CompleteResourceTokenAuth` API를 호출합니다
- **이 튜토리얼에서**: `oauth2_callback_server.py`가 이를 자동으로 처리합니다

#### 5단계: OAuth 플로우 실행
- 사용자가 agent 상호 작용을 통해 OAuth 플로우를 트리거합니다
- 사용자가 권한 부여를 위해 외부 provider로 리디렉션됩니다
- Provider가 세션 정보와 함께 콜백으로 다시 리디렉션합니다
- Session binding이 완료되고 OAuth 토큰을 사용할 수 있게 됩니다

### 환경 인식 OAuth2 Callback Server

이 튜토리얼은 다양한 실행 환경에 자동으로 적응하는 환경 인식 `oauth2_callback_server.py`를 사용합니다:

#### **로컬 개발:**
- **External Callback URL**: `http://localhost:9090/oauth2/callback` (브라우저 액세스 가능)
- **Internal Communication**: `http://localhost:9090` (notebook ↔ server)
- **Server Binding**: `127.0.0.1` (localhost만, 안전함)

#### **SageMaker Workshop Studio:**
- **External Callback URL**: `https://<domain>.studio.<region>.sagemaker.aws/proxy/9090/oauth2/callback` (프록시를 통해 브라우저 액세스 가능)
- **Internal Communication**: `http://localhost:9090` (동일한 컨테이너의 notebook ↔ server)
- **Server Binding**: `0.0.0.0` (SageMaker 프록시가 서버에 도달할 수 있도록 허용)

OAuth2 callback server는 `/opt/ml/metadata/resource-metadata.json`을 확인하여 환경을 자동으로 감지하고 그에 따라 자체 구성합니다.

### oauth2_callback_server.py의 기능

1. **로컬 FastAPI Server 실행** (포트 9090)
   - 상태 확인을 위한 `/ping` 엔드포인트 제공
   - 사용자 토큰을 저장하기 위한 `/userIdentifier/token` 엔드포인트 제공
   - OAuth 리디렉션을 처리하기 위한 `/oauth2/callback` 엔드포인트 제공

2. **사용자 Token 저장 관리**
   - Cognito 인증에서 사용자의 JWT 토큰을 저장합니다
   - OAuth 세션을 올바른 사용자 ID와 연결합니다

3. **OAuth Callback 처리**
   - `session_id` 매개변수와 함께 OAuth 리디렉션을 수신합니다
   - `CompleteResourceTokenAuth`를 호출하여 세션을 바인딩합니다
   - 플로우를 완료하기 전에 사용자 ID를 검증합니다

4. **Session 보안 제공**
   - OAuth 세션이 인증된 사용자에게 바인딩되도록 합니다
   - OAuth 토큰에 대한 무단 액세스를 방지합니다

5. **환경 감지**
   - 로컬 vs SageMaker Studio 환경을 자동으로 감지합니다
   - URL 및 서버 바인딩을 적절하게 구성합니다

```python
    workload_name = launch_result.agent_id
    workload_identity = identity_client.get_workload_identity(name=workload_name)
    allowed_resource_oauth_2_return_urls = workload_identity.get("allowedResourceOauth2ReturnUrls") or []
    oauth2_callback_url = get_oauth2_callback_url()
    print(f"Updating workload {workload_name} with callback url {oauth2_callback_url}")

    updated_workload_identity = identity_client.update_workload_identity(
        name=workload_name,
        allowed_resource_oauth_2_return_urls=[*allowed_resource_oauth_2_return_urls, oauth2_callback_url],
    )
```

이 코드는:
1. **Agent의 Workload Identity 검색**: runtime deployment에서 agent ID를 사용합니다
2. **현재 허용된 URL 가져오기**: 기존 `allowedResourceOauth2ReturnUrls`를 가져옵니다
3. **환경 인식 Callback URL 추가**: 
   - 로컬: `http://localhost:9090/oauth2/callback`
   - SageMaker: `https://<domain>.studio.sagemaker.aws/proxy/9090/oauth2/callback`
4. **Workload Identity 업데이트**: AgentCore Identity에 콜백 URL을 등록합니다

이 등록은 AgentCore Identity가 사전 등록된 URL로만 OAuth 콜백을 리디렉션하므로 추가 보안 계층을 제공하기 때문에 필수적입니다.

### 보안 고려 사항

OAuth2 session binding 프로세스에는 여러 보안 조치가 포함됩니다:
- **URL 검증**: 사전 등록된 콜백 URL만 허용됩니다
- **Session 확인**: 토큰 완료 전에 사용자 세션을 검증해야 합니다
- **사용자 Identity Binding**: OAuth 세션이 인증된 사용자에게 명시적으로 바인딩됩니다
- **Token 격리**: 각 사용자의 OAuth 토큰이 격리되고 안전합니다
- **환경 인식 URL**: 각 환경에 적합한 URL을 자동으로 사용합니다

이 포괄적인 접근 방식은 로컬 또는 SageMaker Workshop Studio에서 실행되는지 여부에 관계없이 다중 사용자 환경에서 OAuth2 플로우가 안전하고 올바른 사용자에게 적절하게 귀속되도록 합니다.

참고: 운영 모범 사례는 코드를 컨테이너로 패키징하고 CI/CD 파이프라인 및 IaC를 사용하여 ECR에 푸시하는 것입니다

### AgentCore Runtime에 agent 시작

이제 docker 파일이 있으므로 agent를 AgentCore Runtime에 시작하겠습니다. 이렇게 하면 Amazon ECR 리포지토리와 AgentCore Runtime이 생성됩니다

````python
from oauth2_callback_server import get_oauth2_callback_url

# Agent를 AgentCore Runtime에 배포하고 배포 정보 가져오기
launch_result = agentcore_runtime.launch(
    env_vars={"CALLBACK_URL": oauth2_callback_url_for_agent},  # OAuth2 callback URL을 환경 변수로 전달
    auto_update_on_conflict=True,  # 충돌 시 자동 업데이트
)
print(launch_result)

if launch_result.agent_id:
    # 배포된 agent의 ID를 workload 이름으로 사용
    workload_name = launch_result.agent_id
    # AgentCore Identity에서 현재 workload identity 구성 가져오기
    workload_identity = identity_client.get_workload_identity(name=workload_name)
    # 이미 등록된 OAuth2 callback URL 목록 추출
    allowed_resource_oauth_2_return_urls = (
        workload_identity.get("allowedResourceOauth2ReturnUrls") or []
    )
    # Session binding을 위한 로컬 OAuth2 callback 서버 URL 가져오기
    oauth2_callback_url = get_oauth2_callback_url()
    print(f"Updating workload {workload_name} with callback url {oauth2_callback_url}")

    # OAuth2 session binding을 활성화하기 위해 workload identity에 로컬 callback URL 등록
    updated_workload_identity = identity_client.update_workload_identity(
        name=workload_name,
        allowedResourceOauth2ReturnUrls=[
            *allowed_resource_oauth_2_return_urls,
            oauth2_callback_url,
        ],
    )
    print(updated_workload_identity)
````

#### 자동 생성된 role에 추가 필수 정책 추가

이전에 model에 액세스하지 않은 새 계정에서 이를 사용하는 경우 agent가 model에 액세스할 수 있도록 자동 생성된 role에 추가 필수 정책을 추가해야 합니다.

````python
import json
import boto3

agentcore_control_client = boto3.client("bedrock-agentcore-control", region_name=region)

# Agent runtime 정보 가져오기
runtime_response = agentcore_control_client.get_agent_runtime(
    agentRuntimeId=launch_result.agent_id
)
runtime_role = runtime_response["roleArn"]
account = boto_session.client("sts").get_caller_identity().get("Account")

# Agent runtime role에 추가할 IAM 정책 정의
policies_to_add = {
    "Version": "2012-10-17",
    "Statement": [
        {
            "Sid": "BedrockModelAccess",
            "Effect": "Allow",
            "Action": [
                "aws-marketplace:ViewSubscriptions",
                "aws-marketplace:Subscribe",
            ],
            "Resource": "*",
        },
        {
            "Sid": "Oauth2TokenAccess",
            "Effect": "Allow",
            "Action": [
                "bedrock-agentcore:GetResourceOauth2Token",  # OAuth2 토큰 가져오기 권한
            ],
            "Resource": "*",
        },
        {
            "Sid": "SecretsManagerAccess",
            "Effect": "Allow",
            "Action": [
                "secretsmanager:GetSecretValue",  # GitHub OAuth2 credential 접근 권한
            ],
            "Resource": [
                f"arn:aws:secretsmanager:{region}:{account}:secret:secret:bedrock-agentcore-identity!default/oauth2/github-provider*"
            ],
        },
    ],
}
iam_client = boto3.client("iam", region_name=region)

# Runtime role에 정책 추가
response = iam_client.put_role_policy(
    PolicyDocument=json.dumps(policies_to_add),
    PolicyName="outbound_policies",
    RoleName=runtime_role.split("/")[1],
)
````

### AgentCore Runtime 호출

마지막으로 페이로드로 AgentCore Runtime을 호출할 수 있습니다

agent가 "inspect_github_repos" tool을 호출하고 3 Legged OAuth 플로우를 트리거하는 것을 확인할 수 있습니다. Authorization Url이 표시됩니다. Authorization url을 클릭하거나 새 브라우저 세션/탭에 복사/붙여넣기하여 사용자 동의 플로우를 완료하십시오.
Authorization이 완료되면 credential provider "github-provider"가 GitHub에서 access_token을 가져와 tool 실행을 완료하여 GitHub 계정에서 비공개 저장소를 가져옵니다.

<div style="text-align:left">
    <img src="images/invoke.png" width="75%"/>
</div>

````python
import subprocess
from oauth2_callback_server import (
    store_token_in_oauth2_callback_server,
    wait_for_oauth2_server_to_be_ready,
)

# Cognito 사용자 재인증하여 bearer token 발급
bearer_token = reauthenticate_user(cognito_config.get("client_id"))

# OAuth2 callback 서버 시작 명령어 구성
oauth2_callback_server_cmd = [
    sys.executable,
    "oauth2_callback_server.py",
    "--region",
    region,
]
oauth2_callback_server_process = subprocess.Popen(oauth2_callback_server_cmd)

try:
    # OAuth2 callback 서버가 준비될 때까지 대기
    successfully_started_oauth2_server = wait_for_oauth2_server_to_be_ready()
    if not successfully_started_oauth2_server:
        print(
            "Failed to start OAuth2 callback server to handle session binding "
            "(https://docs.aws.amazon.com/bedrock-agentcore/latest/devguide/oauth2-authorization-url-session-binding.html)"
        )
    else:
        # OAuth2 callback 서버에 사용자 토큰 저장 (session binding용)
        store_token_in_oauth2_callback_server(bearer_token)
        # Agent 호출 (GitHub 비공개 저장소 조회)
        invoke_response = agentcore_runtime.invoke(
            {"prompt": "What are my private repositories?"}, bearer_token=bearer_token
        )
        print(invoke_response)
finally:
    # OAuth2 callback 서버 프로세스 종료
    oauth2_callback_server_process.terminate()
````
